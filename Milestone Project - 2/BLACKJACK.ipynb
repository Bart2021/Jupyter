```python
import random

clubs = []
diamonds = []
hearts = []
spades = []

dealer_cards = []
player_cards = []


class Card():
    
    def __init__(self,shape,value_str,value):
        
        self.shape = shape
        self.value_str = value_str
        self.value = value
        if value == 1:
            self.points = 11
        elif value > 1 and value < 11:
            self.points = value
        else:
            self.points = 10
            
    def __str__(self):
        
        return f'{self.value_str} of {self.shape}'





        
def card_draw_from_deck():
    
    while True:
        
        shape = random.choice(['Clubs','Diamonds','Hearts','Spades'])
        value = random.randint(1, 13)

        if shape == clubs[0]:
            if clubs[value] != 'taken':

                karta = Card(clubs[0],clubs[value],value)
                clubs[value] = 'taken'
                return karta
            else:
                continue
                
        elif shape == diamonds[0]:
            if diamonds[value] != 'taken':

                karta = Card(diamonds[0],diamonds[value],value)
                diamonds[value] = 'taken'
                return karta
            else:
                continue
                
        elif shape == hearts[0]:
            if hearts[value] != 'taken':

                karta = Card(hearts[0],hearts[value],value)
                hearts[value] = 'taken'
                return karta
            else:
                continue
                
        elif shape == spades[0]:
            if spades[value] != 'taken':

                karta = Card(spades[0],spades[value],value)
                spades[value] = 'taken'
                return karta
            else:
                continue
                
                
                
def player_wins():
    suma = 0
    for num in range(0,len(player_cards)):
        suma += player_cards[num].points
    return suma == 21


def player_busts():
    suma = 0
    for num in range(0,len(player_cards)):
        suma += player_cards[num].points
    return suma > 21




# ta pętla odpowiada za liczbę partyjek
while True:
    
    clubs = ['Clubs','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    diamonds = ['Diamonds','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    hearts = ['Hearts','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    spades = ['Spades','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    
    bet = 0
    
    # the hand begins - partyjka się zaczyna
    print('Welcome to BlackJack!\n')
    
    # asking for the amount of chips
    while True:
        try:
            bet = int(input('How many chips would you like to bet?  '))
            
        except:
            print('Please provide a number!\n')
            
        else:
            break
            
    
    # początkowe rozdanie kart---------------------------
    dealer_cards = []
    player_cards = []
    
    player_cards.append(card_draw_from_deck())
    player_cards.append(card_draw_from_deck())
    
    dealer_cards.append(card_draw_from_deck())
    dealer_cards.append(card_draw_from_deck())
    
    print("\nDealer's Hand:")
    print(' <card hidden>')
    print(f' {dealer_cards[1]}')
    
    print("\nPlayer's Hand:")
    print(f' {player_cards[0]}')
    print(f' {player_cards[1]}')
    
    
    # PLAYER TURN @@@@@@@@@@@@@@@@@@@@@@@@@@@
    while True:
        if player_wins():
            print('\nBlackjack! Player wins!')
            print(f"Player's winnings stand at {bet*1.5}")
            break
            
        # we're asking for the next move
        while True:
            move = input("\nWould you like to Hit or Stand? Enter 'h' or 's'  ")
            if move == 'h' or move == 's':
                break
                
        # player pressed 'h'
        if move == 'h':
            player_cards.append(card_draw_from_deck())
            # we display the cards
            print('----------------')
            print("Dealer's Hand:")
            print(' <card hidden>')
            print(f' {dealer_cards[1]}')
            
            print("\nPlayer's Hand:")
            for num in range(0,len(player_cards)):
                print(f' {player_cards[num]}')
                
            #we check if player busts
            if player_busts():
                print('\nBust! Player lost!')
                break
            
            
        else:
            print('Player stands. Dealer is playing.')
            # we display the cards
            print('----------------')
            print("Dealer's Hand:")
            print(' <card hidden>')
            print(f' {dealer_cards[1]}')
            
            print("\nPlayer's Hand:")
            for num in range(0,len(player_cards)):
                print(f' {player_cards[num]}')
            break
        
        
    punkty_player = 0
    for num in range(0,len(player_cards)):
        punkty_player += player_cards[num].points
    
    
    punkty_dealer = 0
    
    # DEALER TURN @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
    while not player_wins() and not player_busts():
        punkty_dealer = 0
        # we display the cards
        print('----------------')
        print("Dealer's Hand:")
        for num in range(0,len(dealer_cards)):
            print(f' {dealer_cards[num]}')
            punkty_dealer += dealer_cards[num].points
            
        print("\nPlayer's Hand:")
        for num in range(0,len(player_cards)):
            print(f' {player_cards[num]}')
                
        if punkty_dealer == 21:
            print('\nBlackjack! Dealer wins!')
            break
            
        if punkty_dealer > 21:
            print('\nBust! Dealer lost!')
            print(f"Player's winnings stand at {2*bet}")
            break
            
        if punkty_dealer > punkty_player:
            print('\nDealer wins!')
            break
            
        if punkty_dealer <= punkty_player:
            dealer_cards.append(card_draw_from_deck())
            
        
    while True:
        partia = input("\nWould you like to play another hand? Enter 'y' or 'n'  ")
        if partia == 'y' or partia == 'n':
            break
            
    if partia == 'n':
        print('Thanks for playing!')
        break
    
```
